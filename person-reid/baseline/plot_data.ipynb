```python
import matplotlib
import matplotlib.pyplot as plt
import os
import cv2
%matplotlib inline
```

```python
train_data_dir="data/Market-1501-v15.09.15/bounding_box_train/"
train_list=os.listdir(train_data_dir)
test_data_dir="data/Market-1501-v15.09.15/bounding_box_test/"
test_list=os.listdir(test_data_dir)
print("total train images {}".format(len(train_list)))
print("total test images {}".format(len(train_list)))
```

```python

for i,image_path in enumerate(train_list[:3]):
    img=cv2.imread(train_data_dir+image_path)
    img=img/255
    plt.subplot(131 + i)
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.title(image_path)
```

```python
for i,image_path in enumerate(train_list[:3]):
    img=cv2.imread(train_data_dir+image_path)
    print(img.shape)
```

```python
for i,image_path in enumerate(test_list[:3]):
    img=cv2.imread(test_data_dir+image_path)
    print(img.shape)
```
